```python
import spacy
nlp = spacy.load("en_core_web_sm")

# Print the default set of stop words from the loaded English model
print(nlp.Defaults.stop_words)
```

This block loads the `en_core_web_sm` English language model from spaCy and then prints its default set of stop words.

```python
# Print the total number of default stop words
print(len(nlp.Defaults.stop_words))

# Check if 'myself' is considered a stop word
print(nlp.vocab["myself"].is_stop)

# Check if 'mystery' is considered a stop word
print(nlp.vocab["mystery"].is_stop)
```

This block first shows the count of default stop words. Then, it demonstrates how to check if a specific word is marked as a stop word in spaCy's vocabulary using `nlp.vocab[word].is_stop`.

```python
# Add 'btw' to the set of default stop words
nlp.Defaults.stop_words.add("btw")

# Mark 'btw' as a stop word in the vocabulary
nlp.vocab["btw"].is_stop = True

# Print the new total number of stop words
print(len(nlp.Defaults.stop_words))

# Verify that 'btw' is now considered a stop word
print(nlp.vocab["btw"].is_stop)
```

This final block shows how to extend the default stop word list. You add the word to `nlp.Defaults.stop_words` and then update its `is_stop` attribute in the `nlp.vocab` to reflect this change. The length and status of 'btw' are then re-checked to confirm the modification.
