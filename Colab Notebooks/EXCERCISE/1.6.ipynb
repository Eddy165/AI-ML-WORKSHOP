```python
import nltk
```

### Porter Stemmer

The Porter Stemmer is one of the most common stemming algorithms. It is known for its simplicity and reasonable performance.

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

Applying Porter Stemmer to a list of words:

```python
words = ['run', 'runner', 'running', 'ran', 'runs', 'easily', 'fairly']
for word in words:
    print(word + ' --> ' + p_stemmer.stem(word))
```

Applying Porter Stemmer to words in a phrase:

```python
phrase = 'I am meeting him tomorrow at the meeting'
for word in phrase.split():
    print(word + ' --> ' + p_stemmer.stem(word))
```

### Snowball Stemmer (or English Stemmer)

The Snowball Stemmer is an improvement over the Porter Stemmer, offering more aggressive stemming and supporting multiple languages.

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
```

Applying Snowball Stemmer to a list of words:

```python
words = ['run', 'runner', 'running', 'ran', 'runs', 'easily', 'fairly']
for word in words:
    print(word + ' --> ' + s_stemmer.stem(word))
```

Applying Porter Stemmer to a different phrase to see its effects on various word forms:

```python
phrase = "The train is always arriving late so I couldn't come to the college early"
for word in phrase.split():
    print(word + ' --> ' + p_stemmer.stem(word))
```
