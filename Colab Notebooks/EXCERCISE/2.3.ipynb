```python
import spacy
from spacy.matcher import Matcher,PhraseMatcher
from spacy import displacy

nlp=spacy.load("en_core_web_sm")
```

```python
matcher=Matcher(nlp.vocab)
pattern1=[{"LOWER":"solarpower"}]
pattern2=[{"LOWER":"solar"},{"LOWER":"power"}]
pattern3=[{"LOWER":"solarpowered"}]
pattern4=[{"LOWER":"solar"},{"IS_PUNCT":True},{"LOWER":"power"}]
matcher.add("SolarPower",[pattern1,pattern2,pattern3,pattern4])
doc=nlp("The Solar Power industry continues to grow as demand for solarpower increases.Solar-power cars are gaining popularity.")
found_matches=matcher(doc)
print(found_matches)
for match_id,start,end in found_matches:
 string_id=nlp.vocab.strings[match_id]
 span=doc[start:end]
 print(match_id,string_id,start,end,span.text)
```

```python
matcher=PhraseMatcher(nlp.vocab)
phrase_list=["voodoo economics","supply-side economics","trickle-down economics","free-market economics"]
phrase_patterns=[nlp(text) for text in phrase_list]
matcher.add("VoodooEconomics",phrase_patterns)
doc=nlp("Reagan promoted supply-side economics also known as trickle-down economics and some called it voodoo economics supporting free-market economics.")
matches=matcher(doc)
print(matches)
```

```python
doc=nlp("The Quick brown fox jumper over the lazy dog's back.")
print(doc.text)
print(doc[4].text,doc[4].pos_,doc[4].tag_,spacy.explain(doc[4].tag_))
```

```python
doc=nlp("I read books on NLP.")
r=doc[1]
print(f"{r.text:{10}}{r.pos_:{8}}{r.tag_:{6}}{spacy.explain(r.tag_)}")
```

```python
doc=nlp("I read a book on NLP.")
r=doc[1]
print(f"{r.text:{10}}{r.pos_:{8}}{r.tag_:{6}}{spacy.explain(r.tag_)}")
Dep_Counts=doc.count_by(spacy.attrs.DEP)
for k,v in sorted(Dep_Counts.items()):
 print(f"{k}.{doc.vocab[k].text:{4}}:{v}")
```

```python
doc=nlp("The Quick brown fox jumper over the lazy dog's back.")
for token in doc:
 print(f"{token.text:{10}}{token.pos_:{2}}{token.dep_:{7}}{spacy.explain(token.dep_)}")
```
