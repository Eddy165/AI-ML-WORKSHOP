```python


with open('Loyola.txt', 'w') as file:
    file.write("Hello, this is a quick test file.\n")
    file.write("This is the second line of the file.")

print("File Created Successfully ")


with open('Loyola.txt', 'r') as my_file:
    print("\nFile Content:")
    print(my_file.read())



with open('Loyola.txt', 'w+') as my_file:
    my_file.write('This is a new first line')

    my_file.seek(0)
    print("\nAfter Overwriting:")
    print(my_file.read())

with open('test.txt', 'a+') as my_file:
    my_file.write('\nThis line is being appended to test.txt')
    my_file.write('\nAnd another line here.')

    my_file.seek(0)
    print("\nAppended File Content:")
    print(my_file.read())


with open('Loyola.txt', 'r') as txt:
    first_line = txt.readline()

print("\nFirst Line Only:")
print(first_line)
```
